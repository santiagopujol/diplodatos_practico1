## 3.2
Realizar una prueba de Kolmogorov-Smirnof para comprobar analíticamente si estas variables responden la distribución propuesta en el ejercicio anterior. Hint: podés usar https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.kstest.html, pero hay que tener en cuenta que si la distribución es "norm", entonces va a comparar los datos con una distribución normal con media 0 y desviación estándar 1. Se puede utilizar la distribución sobre todos los datos o sólo sobre Latinoamérica.

```python
# Normalizamos para pf_identity
mean_pf, std_pf = ss.norm.fit(dataset['pf_identity'].dropna())
norm_pf = (dataset['pf_identity'].dropna() - mean_pf) / std_pf

# Normalizamos para hf_score
mean_hf, std_hf = ss.norm.fit(dataset['hf_score'].dropna())
norm_hf = (dataset['hf_score'].dropna() - mean_hf) / std_hf

# Obtenemos los indicadores para pf_identity
kstest_pf = ss.kstest(numpy.array(norm_pf), 'norm')

# Obtenemos los indicadores para hf_score
kstest_hf = ss.kstest(numpy.array(norm_hf), 'norm')

print('pvalue_pf', kstest_pf, ' - pvalue_hf', kstest_hf)
```

```python
## Dado que para ambas variables los pvalue son próximos a cero no podemos aseverar 
## que las distribuciones en ambos casos sean normales.
```

## 3.3
Realizar un gráfico QQ de las mismas distribuciones. Se puede utilizar a,bas distribuciones sobre todos los datos o sólo sobre Latinoamérica, pero no cruzadas.

```python
percs = numpy.linspace(0,100,21)  # Creamos 20 puntos percentiles igualmente distribuidos entre 0 y 100.
qn_pf_identity = numpy.percentile(dataset['pf_identity'].dropna(), percs)
qn_hf_score = numpy.percentile(dataset['hf_score'].dropna(), percs)
plt.figure(figsize=(12,6))
seaborn.regplot(x=qn_pf_identity, y=qn_hf_score)
plt.xlabel('Percentiles de pf_identity a nivel Global')
plt.ylabel('Percentiles de pf_identity a nivel Global')

# Graficamos la linea de la identidad.
x = numpy.linspace(numpy.min((qn_pf_identity.min(),qn_hf_score.min())), numpy.max((qn_pf_identity.max(),qn_hf_score.max())))
plt.plot(x,x, color=RED, ls="--")

seaborn.despine()
```

```python
## El gráfico muestra una correlación lineal entre ambas distribuciones pero no lo suficiente para decir que son iguales.
```

## 3.4
Medir la asimetría (skew) y curtosis de las mismas distribuciones con las que realizó el gráfico anterior. ¿Cómo se relacionan estos estadísticos con la forma del gráfico QQ obtenido previamente? ¿El gráfico QQ provee más información que no esté presente en estos estadísticos?

```python
# Valor de asimetría pf_identity nivel Global
skew_pf_identity = ss.skew(dataset['pf_identity'].dropna())
# Valor de asimetría hf_score nivel Global
skew_hf_score = ss.skew(dataset['hf_score'].dropna())
# Valor de curtosis pf_identity nivel Global
kurt_pf_identity = ss.kurtosis(dataset['pf_identity'].dropna())
# Valor de curtosis hf_score nivel Global 
kurt_hf_score = ss.kurtosis(dataset['hf_score'].dropna())

print('Asimetría para pf_identity = {0}'.format(skew_pf_identity))
print('Asimetría para hf_score = {0}'.format(skew_hf_score))
print('Curtosis para pf_identity = {0}'.format(kurt_pf_identity))
print('Curtosis para hf_score = {0}'.format(kurt_hf_score))
```

```python
# Un valor negativo para la asimetría indica una distribución alargada hacia la izquierda, 
# hacia valores menores a la media. Esto se observa para ambas (pf_identity y hf_score).
# El valor negativo para la curtosis indica una distribución algo aplanada, esto significa una mayor dispersión de los datos 
# respecto de la media. Esto se observa para ambas (pf_identity y hf_score).
```
